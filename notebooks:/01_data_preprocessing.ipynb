````python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 4))

# 畫 boxplot，異常值設定為「空心橄欖綠圓圈」
box = plt.boxplot(
    df['BMI'],
    patch_artist=True,
     vert=False,
    flierprops=dict(
        marker='o',
        markerfacecolor='none',         # 不填滿異常值
        markeredgecolor='#007231',      # 異常值邊框為橄欖綠
        markersize=6
    )
)

# 空心盒子 + 國泰綠邊線
for patch in box['boxes']:
    patch.set_facecolor('none')              # 不填色
    patch.set_edgecolor('#B0B0B0')           # 線條為 #5d8736
    patch.set_linewidth(2)

# 鬚鬚與上下限為淺灰色
for whisker in box['whiskers']:
    whisker.set(color='#B0B0B0', linewidth=1.5)
for cap in box['caps']:
    cap.set(color='#B0B0B0', linewidth=1.5)

# 中位數線條為深綠
for median in box['medians']:
    median.set(color='#B0B0B0', linewidth=2)

# 軸線也用淺灰色
for spine in plt.gca().spines.values():
    spine.set_color('#B0B0B0')

# 座標軸與標題顏色
plt.title("BMI", color='#5d8736', fontsize=16)
plt.tick_params(axis='y', colors='#B0B0B0')
plt.tick_params(axis='x', colors='#B0B0B0')

plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
````

````python
from scipy.stats import ttest_ind

# 取出兩組資料
bmi_yes = df_clean[df_clean['HadHeartAttack'] == 'Yes']['BMI']
bmi_no = df_clean[df_clean['HadHeartAttack'] == 'No']['BMI']

# 執行 Welch's t-test（不假設變異數相等，較穩健）
t_stat, p_value = ttest_ind(bmi_yes, bmi_no, equal_var=False)

print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("✅ 結論：兩組之間的 BMI 有顯著差異（p < 0.05）")
else:
    print("❌ 結論：兩組之間的 BMI 無顯著差異（p ≥ 0.05）")
````

````python
from scipy.stats import ttest_ind

# 取出兩組資料
bmi_yes = df[df['HadHeartAttack'] == 'Yes']['BMI']
bmi_no = df[df['HadHeartAttack'] == 'No']['BMI']

# 執行 Welch's t-test（不假設變異數相等，較穩健）
t_stat, p_value = ttest_ind(bmi_yes, bmi_no, equal_var=False)

print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("✅ 結論：兩組之間的 BMI 有顯著差異（p < 0.05）")
else:
    print("❌ 結論：兩組之間的 BMI 無顯著差異（p ≥ 0.05）")
````
