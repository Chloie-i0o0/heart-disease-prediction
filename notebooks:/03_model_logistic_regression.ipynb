```python
import statsmodels.api as sm

def backward_stepwise_selection(X, y, threshold_out=0.05, verbose=True):
    """
    使用 backward elimination 做變數選擇。
    每次迭代移除 p-value > threshold_out 的變數。
    
    參數：
    - X: 資料框（DataFrame），自變數
    - y: Series，目標變數
    - threshold_out: 判定變數是否移除的 p 值門檻
    - verbose: 是否顯示每一步的移除變數
    
    回傳：
    - selected_vars: 被保留下來的變數名稱列表
    - final_model: 最終 statsmodels 擬合模型
    """
    X = sm.add_constant(X.copy())
    remaining_vars = list(X.columns)
    removed = True

    while removed:
        removed = False
        model = sm.Logit(y, X[remaining_vars]).fit(disp=0)
        pvalues = model.pvalues.drop('const', errors='ignore')  # 不考慮常數項

        worst_pval = pvalues.max()
        if worst_pval > threshold_out:
            worst_feature = pvalues.idxmax()
            remaining_vars.remove(worst_feature)
            removed = True
            if verbose:
                print(f"❌ 移除變數：{worst_feature}（p = {worst_pval:.4f}）")

    final_model = sm.Logit(y, X[remaining_vars]).fit(disp=0)
    print("✅ 最佳模型變數：", remaining_vars)
    return remaining_vars, final_model
```

```python
# threshold

import numpy as np
import pandas as pd
from sklearn.metrics import precision_score, recall_score, f1_score, fbeta_score, roc_auc_score

def evaluate_thresholds(y_true, y_proba, thresholds=None):
    if thresholds is None:
        thresholds = np.arange(0.1, 0.9, 0.05)
    rows = []
    for t in thresholds:
        y_pred = (y_proba >= t).astype(int)
        rows.append({
            'Threshold': round(t, 2),
            'Precision': precision_score(y_true, y_pred),
            'Recall': recall_score(y_true, y_pred),
            'F1': f1_score(y_true, y_pred),
            'F2': fbeta_score(y_true, y_pred, beta=2),
            'ROC-AUC': roc_auc_score(y_true, y_proba)
        })
    return pd.DataFrame(rows)

# 假設你已經有：
#   y_test, y_proba_test (從 pipe_final.predict_proba 得到)
threshold_df = evaluate_thresholds(y_test, y_proba_test, thresholds=np.arange(0.3, 0.7, 0.01))
threshold_df = threshold_df.sort_values('F2', ascending=False).reset_index(drop=True)

print(threshold_df)
```
